````python
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
````

````python
sns.jointplot(x="Age", y="Fare", data=df_train, kind='scatter')
````

````python
sns.jointplot(x="Age", y="Fare", data=df_train, kind='kde')
````

````python
sns.jointplot(x="Age", y="Fare", data=df_train, kind='resid')
````

````python
sns.jointplot(x="Age", y="Fare", data=df_train, kind='hex')
````
